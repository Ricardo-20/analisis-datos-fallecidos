# CARGAR DATOS DEL ARCHIVO FALLECIDOS_SINADEF

### Antes de cargar los datos directamente del archivo "fallecidos_sinadef.csv", se realiza un data cleaning de los datos que no seran necesarios para este ejercicio, seguidamente creamos otro archivo .csv que contenga los datos principales que usaremos.

#### Importamos la libreria pandas

```python
import pandas as pd
```

#### Cargamos el archivo "fallecidos_data_clean.csv"

```python
df=pd.read_csv("fallecidos_data_clean.csv",  delimiter=";", encoding= 'unicode_escape')
df.head()
```

#### Verificamos que tipos de datos tenemos en nuestro DataFrame

```python
df.dtypes
```

#### Despues de realizar un detenido análisis, podemos observar que nuestra columna 'EDAD' es de tipo object lo cual nos dificulta hallar la media respectiva de las edades.
*Es necesario que los datos de la columna 'EDAD' sea de tipo INT, por eso que se averigua el motivo en que la columna 'EDAD' es de tipo object y tenemos como resultado, que es a causa de la variable 'SIN REGISTRO', afectando y obligando que la columna sea de tipo object.*

```python
df[df.EDAD =='SIN REGISTRO' ]
```

#### Antes de continuar con el cambio de tipo de datos en la columna 'EDAD', es necesario saber si todas las filas del dataframe se encuentran completas sin datos en blanco, debido a que esto puede afectar en el resultado cuando vamos a hallar nuestra media.

```python
df.isnull().any()
```

**Concluimos que nuestro dataframe se encuentra libre de datos nulos de manera automatica, pero en vez de obtener datos nulos estas han sido rellenadas con el nombre de "SIN REGISTRO", por lo que procedemos a continuar con la eliminación de las filas que contienen este tipo de variable antes de realizar el cambio de tipo de datos de la columna**

*Tenemos un total de 2200 filas de todo nuestro dataframe con una variable 'SIN REGISTRO'*

```python
df[df.EDAD =='SIN REGISTRO'].count().sum()
```

### Número de filas totales incluidas las que estan sin registrar la edad de los fallecidos

```python
df.count()
```

### Eliminando todas las filas de los fallecidos que estan sin registrar la edad

```python
df = df.drop(df[df.EDAD == 'SIN REGISTRO'].index)
df.head()
```

### Comprobación de que se eliminaron los fallecidos sin registrar la edad

```python
df[df.EDAD =='SIN REGISTRO'].count().sum()
```

**Debido a que eliminamos la variable 'SIN REGISTRAR', que obligaba a la columna 'EDAD' que sea de tipo object, ahora se podra cambiar el tipo de dato a int de la columna para hallar la media.**

```python
df['EDAD']=df['EDAD'].astype(str).astype(int)
df.dtypes
```

### Hallamos todos los datos estadísticos básicos de la columna 'EDAD'

```python
df.describe()
```

### Finalmente hallamos la media de las edades de las personas fallecidas

```python
df['EDAD'].mean()
```

### Podemos observar que en la columna 'EDAD' el número máximo es de 3017 y esto se debe que dentro de nuestra columna 'TIEMPO EDAD', existe una variable de tipo HORAS.
**Las edades que estan relacionadas con el tiempo de edad en horas, podemos considerar a gran escala como si fuera 1 año de vida, esto es debido al gran volumen de edades que esta siendo manejado en años.**

```python
#De manera lógica es imposible en estos tiempos que una persona sobrepase los 149 años de edad
df[df.EDAD >= 149].count().sum()
```

```python
df[(df['TIEMPO EDAD']=="HORAS")&(df.EDAD>1)]
```

### Eliminando todas las filas de los fallecidos que estan con horas dentro de la columna tiempo de edad y sus horas de vida ya que estas equivalen a 1 año únicamente para este análisis de datos

```python
df = df.drop(df[(df['TIEMPO EDAD']=="HORAS")&(df.EDAD>1)].index)
df.head()
```

### Comprobación de que se eliminaron los fallecidos con horas de vida

```python
df[(df['TIEMPO EDAD']=="HORAS")&(df.EDAD>1)].count().sum()
```

### De igual manera dentro de nuestra columna 'TIEMPO EDAD', existen datos que no han sido registrados estos se tendran que eliminar.

```python
df = df.drop(df[df['TIEMPO EDAD'] =='SIN REGISTRO'].index)
df.head()
```

### Comprobación de que se eliminaron los fallecidos sin registrar su tiempo de edad

```python
df[df['TIEMPO EDAD'] =='SIN REGISTRO' ].count().sum()
```

### Volvemos a realizar una comprobación de nuestros dataset para verificar el número máximo de edad
**La edad máxima fue de 148 años**

```python
df.describe()
```

### Media en años según la edad de los fallecidos_sinadef

```python
df['EDAD'].mean()
```
